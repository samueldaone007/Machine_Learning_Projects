Importing the Libraries neccesary for importing the dataset and encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('StudentsPerformance.csv')
```

Viewing the dataset to locate our target column

From the data we decide we want to use the math score as the target for building our model

```python
data.head()
```

Storing the encoder in the variable lab to make the Encoder easy to call and reduce redundancy

```python
lab = LabelEncoder()
```

From here below we are going to be encoding each column step by step

```python
data['gender'] = lab.fit_transform(data['gender'])
```

```python
data.head()
```

```python
data.value_counts('race/ethnicity')
```

```python
data['race/ethnicity'] = lab.fit_transform(data['race/ethnicity'])
```

```python
data.head()
```

```python
data.value_counts('parental level of education')
```

```python
data['parental level of education'] = lab.fit_transform(data['parental level of education'])
```

```python
data.head()
```

```python
data.value_counts('lunch')
```

```python
data['lunch'] = lab.fit_transform(data['lunch'])
```

```python
data
```

```python
data.value_counts('test preparation course')
```

```python
data['test preparation course'] = lab.fit_transform(data['test preparation course'])
```

Now that we are done encoding we will import our algorithms for training and the train_test_split library with will help us split our dataset into training and testing parts

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

Seperating our independent variables from the dependent ones

```python
x = data.drop('math score', axis=1)
```

```python
y = data['math score']
```

```python
x
```

Storing our algorithm in a variable for ease of use

```python
model = RandomForestRegressor()
```

Calling the train test split to split our dataset into training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.1,random_state=42)
```

Training our model

```python
model.fit(x_train,y_train)
```

Checking how well our Model did

As we can see it trained very well

```python
model.score(x_train,y_train)
```

Having our model predict scores based on the independent variables


```python
pred = model.predict(x_test)
```

Our predictions are in decimals and we want them to be in whole numbers to match our dataset

```python
pred
```

We will import numpy and round our figures up or down depending on the figure given

```python
import numpy as np

pred1 = np.round(pred).astype(int)
```

```python
pred1
```

Importing the required libraries to check the score of our regression task

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mae = mean_absolute_error(y_test, pred1)
r2 = r2_score(y_test, pred)*100
```

As we can see our model did well above fairly

```python
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R2): {r2}")
```

Exporting our model for future use

```python
import joblib
```

```python
joblib.dump(model, 'Student_Performance_Predictor.pkl')
```
